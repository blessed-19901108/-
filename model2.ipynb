## 3.2 无造假样本行业建模 - 隔离森林异常检测

### 可视化 - 混淆矩阵 - 异常分数

```python
# 财务造假预测 - 隔离森林建模

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc, precision_recall_curve
from sklearn.model_selection import train_test_split
import os
import warnings
warnings.filterwarnings('ignore')

# 设置中文显示
plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False    # 用来正常显示负号
sns.set(font='SimHei')  # 设置seaborn默认字体

# 1. 加载数据
print("1. 数据加载")
# 加载财务数据
finance_data = pd.read_csv('output/data_standardization.csv')
# 加载行业特征数据
industry_features = pd.read_csv('industry_features_results/all_industries_features.csv')

# 查看数据基本信息
print(f"财务数据形状: {finance_data.shape}")
print(f"行业特征数据形状: {industry_features.shape}")

# 查看财务数据的前几行
print("\n财务数据预览:")
display(finance_data.head())

# 查看行业特征数据的前几行
print("\n行业特征数据预览:")
display(industry_features.head())

# 2. 数据预处理
print("\n2. 数据预处理")


# 检查行业分布
print("\n行业分布:")
industry_count = finance_data['Industry'].value_counts()
display(industry_count)

# 移除"居民服务、修理和其他服务业"行业数据（数据太少）
print("\n移除'居民服务、修理和其他服务业'行业数据")
finance_data = finance_data[finance_data['Industry'] != '居民服务、修理和其他服务业']

# 根据FLAG将数据分为训练集和预测集
training_data = finance_data[finance_data['FLAG'].isin([0, 1])].copy()
prediction_data = finance_data[finance_data['FLAG'].isna()].copy()

print(f"训练数据集形状: {training_data.shape}")
print(f"预测数据集形状: {prediction_data.shape}")

# 3. 定义辅助函数
print("\n3. 定义辅助函数")

def prepare_data_for_industry(df, industry_name, features_df):
    """为特定行业准备数据"""
    industry_data = df[df['Industry'] == industry_name].copy()
    
    # 获取该行业的特征列
    industry_features_list = features_df[features_df['Industry'] == industry_name]['Feature'].tolist()
    
    # 检查特征是否存在于数据中
    valid_features = [feature for feature in industry_features_list if feature in industry_data.columns]
    
    if not valid_features:
        print(f"警告: {industry_name} 没有有效特征，跳过此行业")
        return None, None, None, None, None
    
    # 提取特征和标签
    X = industry_data[valid_features].copy()
    y = industry_data['FLAG'].copy() if 'FLAG' in industry_data.columns else None
    
    # 标准化特征
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    
    # 确保股票代码转换为字符串类型
    ticker_symbols = industry_data['TICKER_SYMBOL'].astype(str).values
    
    return X_scaled, y, valid_features, scaler, ticker_symbols

def evaluate_model(y_true, y_pred, y_scores=None, threshold=0):
    """评估模型性能"""
    # 确保标签格式正确
    y_true = np.array(y_true).ravel()
    y_pred = np.array(y_pred).ravel()
    
    # 混淆矩阵
    conf_mat = confusion_matrix(y_true, y_pred)
    
    # 分类报告
    class_report = classification_report(y_true, y_pred, zero_division=0)
    
    # 计算精确率、召回率、F1分数
    # 处理混淆矩阵维度问题
    if conf_mat.size == 1:  # 只有一个类别的情况
        # 如果只有一个类别，设置默认值
        tn, fp, fn, tp = 0, 0, 0, 0
        if y_true[0] == y_pred[0]:  # 如果唯一的预测正确
            if y_true[0] == 1:  # 如果是正类
                tp = conf_mat[0, 0]
            else:  # 如果是负类
                tn = conf_mat[0, 0]
    elif conf_mat.shape == (1, 1):  # 另一种单类情况
        if 1 in y_true or 1 in y_pred:  # 如果存在正类
            tp = conf_mat[0, 0]
            fn, fp, tn = 0, 0, 0
        else:  # 如果都是负类
            tn = conf_mat[0, 0]
            tp, fn, fp = 0, 0, 0
    elif conf_mat.shape == (2, 2):  # 正常情况
        tn, fp, fn, tp = conf_mat.ravel()
    else:  # 其他情况
        print(f"警告: 混淆矩阵形状异常 {conf_mat.shape}")
        # 使用精确率和召回率函数计算
        from sklearn.metrics import precision_score, recall_score, f1_score
        precision = precision_score(y_true, y_pred, zero_division=0)
        recall = recall_score(y_true, y_pred, zero_division=0)
        f1 = f1_score(y_true, y_pred, zero_division=0)
        
        results = {
            'confusion_matrix': conf_mat,
            'classification_report': class_report,
            'precision': precision,
            'recall': recall,
            'f1_score': f1
        }
        
        return results
    
    # 正常情况下计算指标
    precision = tp / (tp + fp) if (tp + fp) > 0 else 0
    recall = tp / (tp + fn) if (tp + fn) > 0 else 0
    f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
    
    results = {
        'confusion_matrix': conf_mat,
        'classification_report': class_report,
        'precision': precision,
        'recall': recall,
        'f1_score': f1
    }
    
    return results

def visualize_results(results, industry_name):
    """可视化评估结果"""
    # 创建图表布局
    fig, ax = plt.subplots(figsize=(10, 8))
    fig.suptitle(f'{industry_name} - 隔离森林评估结果', fontsize=16)
    
    # 混淆矩阵
    cm = results['confusion_matrix']
    # 检查混淆矩阵的形状
    if cm.shape == (2, 2):
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['正常', '造假'], 
                    yticklabels=['正常', '造假'], ax=ax)
    else:
        # 处理非标准形状的混淆矩阵
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax)
        ax.set_title(f'混淆矩阵 (形状: {cm.shape})')
        
    ax.set_title('混淆矩阵')
    ax.set_xlabel('预测标签')
    ax.set_ylabel('真实标签')
    
    plt.tight_layout()
    plt.savefig(f'plots/{industry_name}_evaluation.png', dpi=600, bbox_inches='tight')
    plt.show()

def analyze_anomaly_score_distribution(scores, industry_name):
    """分析异常分数分布"""
    plt.figure(figsize=(12, 6))
    
    # 创建主图和箱线图子图
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 7))
    
    # 1. 直方图和核密度估计
    sns.histplot(scores, bins=30, kde=True, color='royalblue', ax=ax1)
    ax1.set_title(f'{industry_name} - 异常分数分布', fontsize=14)
    ax1.set_xlabel('异常分数')
    ax1.set_ylabel('频率')
    
    # 添加均值和中位数线
    mean_score = np.mean(scores)
    median_score = np.median(scores)
    ax1.axvline(x=mean_score, color='red', linestyle='--', label=f'均值: {mean_score:.4f}')
    ax1.axvline(x=median_score, color='green', linestyle='-.', label=f'中位数: {median_score:.4f}')
    
    # 计算分位数
    q1, q3 = np.percentile(scores, [25, 75])
    iqr = q3 - q1
    upper_bound = q3 + 1.5 * iqr
    lower_bound = q1 - 1.5 * iqr
    
    # 添加IQR边界线
    ax1.axvline(x=upper_bound, color='purple', linestyle=':', 
              label=f'上界 (Q3+1.5*IQR): {upper_bound:.4f}')
    
    # 标记可能的异常区域
    ax1.axvspan(upper_bound, max(scores) + 0.1, alpha=0.2, color='red', 
              label='潜在异常区域')
    
    ax1.legend(fontsize=10)
    
    # 2. 箱线图和分布图
    sns.boxplot(y=scores, color='lightblue', width=0.3, ax=ax2)
    sns.stripplot(y=scores, color='navy', alpha=0.5, ax=ax2)
    ax2.set_title(f'{industry_name} - 异常分数箱线图', fontsize=14)
    ax2.set_ylabel('异常分数')
    ax2.set_yticks(np.arange(0, max(scores) + 0.2, 0.2))
    ax2.grid(axis='y', linestyle='--', alpha=0.7)
    
    # 添加统计信息文本框
    stats_text = (
        f"统计摘要:\n"
        f"样本数: {len(scores)}\n"
        f"均值: {mean_score:.4f}\n"
        f"中位数: {median_score:.4f}\n"
        f"标准差: {np.std(scores):.4f}\n"
        f"最小值: {min(scores):.4f}\n"
        f"最大值: {max(scores):.4f}\n"
        f"Q1 (25%): {q1:.4f}\n"
        f"Q3 (75%): {q3:.4f}\n"
        f"IQR: {iqr:.4f}\n"
        f"潜在异常阈值: {upper_bound:.4f}"
    )
    
    # 将统计信息放在右侧子图中
    ax2.text(1.05, 0.5, stats_text, transform=ax2.transAxes, 
           bbox=dict(facecolor='lightgrey', alpha=0.5, boxstyle='round,pad=0.5'),
           verticalalignment='center', fontsize=10)
    
    plt.tight_layout()
    plt.savefig(f'plots/{industry_name}_anomaly_score_analysis.png', dpi=600, bbox_inches='tight')
    plt.show()
    
    return {
        'mean': mean_score,
        'median': median_score,
        'std': np.std(scores),
        'min': min(scores),
        'max': max(scores),
        'q1': q1,
        'q3': q3,
        'iqr': iqr,
        'upper_bound': upper_bound,
        'potential_outliers': sum(scores > upper_bound),
        'percent_outliers': sum(scores > upper_bound) / len(scores) * 100
    }

# 4. 对每个行业建立隔离森林模型
print("\n4. 建立隔离森林模型")

# 获取无监督方法的行业
unsupervised_industries = industry_features[industry_features['Method'] == '无监督']['Industry'].unique()
print(f"使用无监督方法的行业数量: {len(unsupervised_industries)}")
print(f"无监督行业列表: {unsupervised_industries}")


# 使用已存在的results和models目录
# 存储预测结果
prediction_results = []
anomaly_stats = []

# 对每个无监督行业建模
for industry in unsupervised_industries:
    print(f"\n处理行业: {industry}")
    
    # 准备训练数据
    X_train, y_train, features, scaler, ticker_symbols = prepare_data_for_industry(
        training_data, industry, industry_features
    )
    
    if X_train is None:
        print(f"跳过行业 {industry}，无有效特征")
        continue
    
    # 获取有FLAG标记的数据
    flagged_data = training_data[(training_data['Industry'] == industry) & 
                                 (training_data['FLAG'].isin([0, 1]))].copy()
    
    # 检查是否有足够的数据用于训练
    if len(flagged_data) < 10:
        print(f"行业 {industry} 的标记数据不足 (N={len(flagged_data)})，跳过评估但仍进行建模")
        evaluation_possible = False
    else:
        evaluation_possible = True
    
    # 对于评估，我们将数据分为训练集和测试集
    if evaluation_possible:
        train_indices, test_indices = train_test_split(
            np.arange(len(flagged_data)), 
            test_size=0.3, 
            stratify=flagged_data['FLAG'],
            random_state=42
        )
        
        # 提取测试集的特征和标签
        test_data = flagged_data.iloc[test_indices].copy()
        X_test = scaler.transform(test_data[features])
        y_test = test_data['FLAG'].values
    
    # 训练隔离森林模型
    print(f"训练隔离森林模型，使用 {len(features)} 个特征")
    isolation_forest = IsolationForest(
        n_estimators=100, 
        max_samples='auto', 
        contamination=0.05,  # 降低异常比例为5%，减少预测为造假的可能性
        random_state=42
    )
    
    # 使用全部训练数据进行拟合
    isolation_forest.fit(X_train)
    
    # 保存模型，以行业名开头
    import joblib
    joblib.dump(isolation_forest, f'models/{industry}_isolation_forest.pkl')
    joblib.dump(scaler, f'models/{industry}_scaler.pkl')
    joblib.dump(features, f'models/{industry}_features.pkl')
    
    # 评估模型
    if evaluation_possible:
        # 在测试集上进行预测
        y_pred_raw = isolation_forest.predict(X_test)
        
        # 隔离森林返回 1 表示正常，-1 表示异常
        # 转换为 0 表示正常，1 表示造假（异常）
        y_pred = (y_pred_raw == -1).astype(int)
        
        # 获取异常分数
        y_scores = -isolation_forest.decision_function(X_test)  # 负值表示更可能是异常
        
        # 评估模型
        results = evaluate_model(y_test, y_pred)
        
        # 打印评估结果
        print(f"\n{industry} 评估结果:")
        print(f"混淆矩阵:\n{results['confusion_matrix']}")
        print(f"\n分类报告:\n{results['classification_report']}")
        
        # 可视化评估结果
        visualize_results(results, industry)
        
        # 分析异常分数分布
        print(f"\n{industry} 异常分数分布分析:")
        stats = analyze_anomaly_score_distribution(y_scores, industry)
        stats['industry'] = industry
        anomaly_stats.append(stats)
        
        # 打印异常分数分析结果
        print(f"潜在异常值比例: {stats['percent_outliers']:.2f}%")
        print(f"异常阈值 (Q3+1.5*IQR): {stats['upper_bound']:.4f}")
    
    # 对预测集中的该行业数据进行预测
    industry_prediction_data = prediction_data[prediction_data['Industry'] == industry].copy()
    
    if len(industry_prediction_data) > 0:
        print(f"为行业 {industry} 的 {len(industry_prediction_data)} 家公司预测")
        
        # 准备预测数据
        pred_features = [f for f in features if f in industry_prediction_data.columns]
        X_pred = industry_prediction_data[pred_features].copy()
        X_pred_scaled = scaler.transform(X_pred)
        
        # 预测
        pred_raw = isolation_forest.predict(X_pred_scaled)
        pred_scores = -isolation_forest.decision_function(X_pred_scaled)
        
        # 转换预测结果
        pred_results = (pred_raw == -1).astype(int)
        
        # 保存预测结果
        for i, idx in enumerate(industry_prediction_data.index):
            prediction_results.append({
                'TICKER_SYMBOL': str(industry_prediction_data.loc[idx, 'TICKER_SYMBOL']),  # 确保股票代码为字符串
                'Industry': industry,
                'Predicted_FLAG': pred_results[i],
                'Anomaly_Score': pred_scores[i]
            })
        
        # 分析预测集的异常分数分布
        if len(pred_scores) > 5:  # 只在有足够样本时分析
            print(f"\n{industry} 预测集异常分数分布分析:")
            pred_stats = analyze_anomaly_score_distribution(pred_scores, f"{industry}_prediction")
            pred_stats['industry'] = f"{industry}_prediction"
            anomaly_stats.append(pred_stats)

# 5. 汇总预测结果和异常分数统计
print("\n5. 汇总预测结果和异常分数统计")
if prediction_results:
    predictions_df = pd.DataFrame(prediction_results)
    print(f"总共为 {len(predictions_df)} 家公司进行了预测")
    
    # 查看预测结果分布
    print("\n预测结果分布:")
    print(predictions_df['Predicted_FLAG'].value_counts())
    
    # 按行业查看预测结果分布
    print("\n按行业查看预测结果分布:")
    industry_pred_counts = predictions_df.groupby(['Industry', 'Predicted_FLAG']).size().unstack(fill_value=0)
    display(industry_pred_counts)
    
    # 保存异常分数统计
    if anomaly_stats:
        anomaly_stats_df = pd.DataFrame(anomaly_stats)
        anomaly_stats_df.to_csv('results/anomaly_score_statistics.csv', index=False)
        print("异常分数统计数据已保存到 'results/anomaly_score_statistics.csv'")
        
        # 比较各行业异常分数分布
        plt.figure(figsize=(14, 8))
        train_stats = anomaly_stats_df[~anomaly_stats_df['industry'].str.contains('_prediction')]
        
        # 创建柱状图比较各行业的异常分数阈值和异常比例
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 8))
        
        # 绘制异常阈值比较
        sns.barplot(x='industry', y='upper_bound', data=train_stats, 
                  palette='Blues_d', ax=ax1)
        ax1.set_title('各行业异常阈值比较', fontsize=14)
        ax1.set_xlabel('行业')
        ax1.set_ylabel('异常阈值 (Q3+1.5*IQR)')
        ax1.tick_params(axis='x', rotation=45)
        for i, v in enumerate(train_stats['upper_bound']):
            ax1.text(i, v + 0.02, f'{v:.3f}', ha='center')
        
        # 绘制异常比例比较
        sns.barplot(x='industry', y='percent_outliers', data=train_stats, 
                  palette='Reds_d', ax=ax2)
        ax2.set_title('各行业潜在异常比例比较', fontsize=14)
        ax2.set_xlabel('行业')
        ax2.set_ylabel('潜在异常比例 (%)')
        ax2.tick_params(axis='x', rotation=45)
        for i, v in enumerate(train_stats['percent_outliers']):
            ax2.text(i, v + 0.5, f'{v:.2f}%', ha='center')
        
        plt.tight_layout()
        plt.savefig('plots/industry_anomaly_comparison.png', dpi=600, bbox_inches='tight')
        plt.show()
    
    # 异常分数分布
    plt.figure(figsize=(10, 6))
    sns.histplot(predictions_df['Anomaly_Score'], bins=50, kde=True, color='royalblue')
    plt.title('异常分数分布')
    plt.xlabel('异常分数')
    plt.ylabel('频率')
    plt.axvline(x=0.5, color='red', linestyle='--', label='阈值: 0.5')
    plt.legend()
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.savefig('plots/anomaly_score_distribution.png', dpi=600, bbox_inches='tight')
    plt.show()
    
    # 按行业可视化异常分数分布
    plt.figure(figsize=(12, 8))
    sns.boxplot(x='Industry', y='Anomaly_Score', data=predictions_df, palette=['lightblue', 'royalblue'])
    plt.title('各行业异常分数分布', fontsize=14)
    plt.xticks(rotation=90)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.savefig('plots/industry_anomaly_scores.png', dpi=600, bbox_inches='tight')
    plt.show()
    
    # 保存预测结果
    predictions_df.to_csv('results/isolation_forest_financial_fraud_predictions.csv', index=False)
    print("预测结果已保存到 'results/isolation_forest_financial_fraud_predictions.csv'")
    
    # 输出各行业预测为造假的股票代码
    print("\n各行业预测为造假的股票代码:")
    fraud_predictions = predictions_df[predictions_df['Predicted_FLAG'] == 1]
    for industry_name in fraud_predictions['Industry'].unique():
        industry_frauds = fraud_predictions[fraud_predictions['Industry'] == industry_name]
        print(f"\n{industry_name} (共 {len(industry_frauds)} 家公司):")
        for i, row in enumerate(industry_frauds.itertuples(), 1):
            print(f"  {i}. 股票代码: {row.TICKER_SYMBOL} (异常分数: {row.Anomaly_Score:.4f})")
    
    # 保存各行业造假股票代码到CSV文件
    fraud_predictions.to_csv('results/isolation_forest_fraud_companies.csv', index=False)
    print("\n预测为造假的公司列表已保存到 'results/isolation_forest_fraud_companies.csv'")
else:
    print("没有生成预测结果")
```
